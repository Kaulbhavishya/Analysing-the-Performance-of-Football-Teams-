```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


La_Liga = pd.read_csv("Laliga.csv")

new_header = La_Liga.iloc[0] #grab the first row for the header
La_Liga = La_Liga[1:] #take the data less the header row
La_Liga.columns = new_header 

La_Liga.replace('-', 0, inplace=True)
La_Liga
```

```python
La_Liga["Pos"] = pd.to_numeric(La_Liga["Pos"])
La_Liga["Seasons"] = pd.to_numeric(La_Liga["Seasons"])
La_Liga["Points"] = pd.to_numeric(La_Liga["Points"])
La_Liga["GamesPlayed"] = pd.to_numeric(La_Liga["GamesPlayed"])
La_Liga["GamesWon"] = pd.to_numeric(La_Liga["GamesWon"])
La_Liga["GamesDrawn"] = pd.to_numeric(La_Liga["GamesDrawn"])
La_Liga["GamesLost"] = pd.to_numeric(La_Liga["GamesLost"])
La_Liga["GoalsFor"] = pd.to_numeric(La_Liga["GoalsFor"])
La_Liga["GoalsAgainst"] = pd.to_numeric(La_Liga["GoalsAgainst"])
La_Liga["Champion"] = pd.to_numeric(La_Liga["Champion"])
La_Liga["Runner-up"] = pd.to_numeric(La_Liga["Runner-up"])
La_Liga["Third"] = pd.to_numeric(La_Liga["Third"])
La_Liga["Fourth"] = pd.to_numeric(La_Liga["Fourth"])
La_Liga["Fifth"] = pd.to_numeric(La_Liga["Fifth"])
La_Liga["Sixth"] = pd.to_numeric(La_Liga["Sixth"])
La_Liga["T"] = pd.to_numeric(La_Liga["T"])
La_Liga["BestPosition"] = pd.to_numeric(La_Liga["BestPosition"])
```
